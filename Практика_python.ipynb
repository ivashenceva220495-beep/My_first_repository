```python
import random

sets_list = []
num_sets = 5  # количество множеств

for i in range(num_sets):
    # Создаём множество из 10 случайных чисел от 0 до 99
    new_set = set(random.sample(range(100), 10))
    sets_list.append(new_set)

# Выводим все множества
for i, s in enumerate(sets_list):
    print(f"Множество {i}: {s}")
```

```python
my_set1 = set()
my_set2 = set()

n = 0
while n <3:
    word1 = input('Введите значение для первого множества: ')
    my_set1.add(word1)
    n+=1
print(my_set1)

y = 0
while y <3:
    word2 = input('Введите значение для второго множества: ')
    my_set2.add(word2)
    y+=1

print(my_set2)

union_set = my_set1.union(my_set2)
print(union_set)

intersection_set = my_set1.intersection(my_set2)
print(intersection_set)

```

```python
# Список списков с парами ключ-значение
nested_pairs = [[("a", 1), ("b", 2)], [("c", 3), ("d", 4)]]

# Генерация словаря из вложенного списка
nested_dict = {key: value for sublist in nested_pairs for key, value in sublist}
print(nested_dict)  # Вывод: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
```

```python
address = {
    "street": "123 Main St",
    "zip": "10001"
}

details = {
    "age": 25,
    "city": "New York",
    "address": address
}

person = {
    "name": "Alice",
    "details": details,
    "gender": "female"
}
print(person)
```

```python
for key, value in person.items():  # Обходим родительский словарь
    if isinstance(value, dict):  # Если значение - это словарь
        for key2, value2 in value.items():  # Обходим элементы дочернего словаря
            print(f"{key} --> {key2}: {value2}")
```
