```python
#находим выбросы через межквартальный размах
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['age'].between(lower_bound, upper_bound)]
remove_outliers
```

```python
# выбросы для столбца age
df[~df['age'].between(lower_bound, upper_bound)]
```

Сделаем замену пропусков средним только у тех столбцов, у которых произошло значительное изменение статистических данных после удаления пропусков : pulse, respiratory rate

```python
#столбец pulse
fill_mean = df.copy()

fill_mean['pulse'] = fill_mean['pulse'].fillna(df['pulse'].mean())
```

```python
#после замены средним у нас есть значительное изменение только в 2 квартиле
print(df['pulse'].describe())
print('-----------------------------------')
print(fill_mean['pulse'].describe())
```

```python
#столбец respiratory rate
fill_mean = df.copy()

fill_mean['respiratory rate'] = fill_mean['respiratory rate'].fillna(df['respiratory rate'].mean())
```

```python
#после замены средним у нас есть значительное изменение только в 2 квартиле
print(df['respiratory rate'].describe())
print('-----------------------------------')
print(fill_mean['respiratory rate'].describe())
```
