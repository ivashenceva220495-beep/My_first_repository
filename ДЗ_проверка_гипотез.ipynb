```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('data_hw_2__upd_.xlsx')
df.head()
```

```python
#Формирование выборок
df_1 = df[df['Country'] == 'France']
df_2 = df[df['Country'] == 'Germany']

print ('Выборки имеют одинаковую длину: ', len(df_1) == len(df_2))
```

Гистограмма плотности объема продаж для Франции и Германии

```python
sns.distplot(df_1['TotalPrice'],50)
sns.distplot(df_2['TotalPrice'],50)
plt.title('Гистограмма плотности объема продаж для Франции и Германии')
plt.legend (['France','Germany'])
plt.show()
```

Подключение библиотек статистических критериев и задание уровня значимости

```python
from scipy.stats import ttest_ind
from scipy.stats import mannwhitneyu

alpha = 0.05 #уровень значимости
```

Критерий Стьюдента


Н0 - Объёмы продаж Франции отличаются от объёмов продаж Германии


Н1 - Объёмы продаж Франции НЕотличаются от объёмов продаж Германии

```python
result = ttest_ind(df_1['TotalPrice'], df_2['TotalPrice'])
print ('p-value: %.8f' % result.pvalue)

if (result.pvalue < alpha):
  print ('Отвергаем нулевую гипотезу')
else:
  print ('НЕ отвергаем нулевую гипотезу')

# H0 - отвергаем гипотезу , следовательно, Объёмы продаж Франции НЕ отличаются от объёмов продаж Германии
```

Проведение теста Манна - Уитни (U - критерий Манна Уитни)

```python
result  = mannwhitneyu(df_1['TotalPrice'], df_2['TotalPrice'])
print ('p-value: %.8f' % result.pvalue)

if (result.pvalue < alpha):
  print ('Отвергаем нулевую гипотезу')
else:
  print ('НЕ отвергаем нулевую гипотезу')

  # H0 - отвергаем гипотезу , следовательно, Объёмы продаж Франции НЕ отличаются от объёмов продаж Германии
```
