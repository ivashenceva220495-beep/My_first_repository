```python
documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
      ]

directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }

#Задание 1. Нахождение владельца документа по номеру документа
def owner_document(document_number):
    for document in documents:
        if document['number'] == document_number:
            print('Владелец документа:' + document['name'])
            break
    else: print('Документ не найден в базе')



#Задание 2. Нахождение полки документа по номеру документа
def get_shelf(document_number):
    for shelf,document_numbers in directories.items():
        if document_number in document_numbers:
            print('Документ хранится на полке: ' + shelf)
            break
    else:
            print('Документ не найден в базе')

#Задание 3. Полная информация по всем документам
def info(documents, directories):
    for shelf, cont in directories.items():
        for doc in documents:
            if doc['number'] in cont:
                print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {shelf}")

#Задание 4. добавление полок.
def add_shelf():

    if shelf in directories.keys():
            print(f"Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}")
    else:
        directories[shelf] = []
        print(f"Полка добавлена. Текущий перечень полок: {list(directories.keys())}")

# Задание 5. Удаление полок.
def rem_shelf(number):
    if number in directories:
        if len(directories(number)) > 0:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + ', '.join(directories.keys()))
        del directories[number]
        print('Полка удалена. Текущий перечень полок: ' + ', '.join(directories.keys()))
    print ('Такой полки не существует. Текущий перечень полок:' + ', '.join(directories.keys()))

while True:
    print('Возможные команды: p, s, l, a,q')
    comand = input('Введите название команды ')
    if comand == 'p':
        document_number = input('Введите номер документа ')
        owner_document(document_number)
    elif comand == 's':
        document_number = input('Введите номер документа ')
        get_shelf(document_number)
    elif comand == 'l':
        info(documents,directories)
    elif comand == 'ads':
        shelf = input("Введите номер полки: ")
        add_shelf()
    elif comand == 'ds':
        number = int(input('Введите номер полки: '))
        rem_shelf(number)
    elif comand == 'q':
        break
    else:
        print ('Неизвестная команда')
```
